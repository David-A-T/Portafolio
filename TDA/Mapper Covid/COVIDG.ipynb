### Preprocesamiento

```python
# Importar: pandas
import pandas as pd
```

```python
# Importar: base de datos a analizar
data = pd.read_csv("Covid_tba2.csv", index_col=0)
data.head()
```

```python
# Transformar: dataframe -> numpy array
data_map = data.to_numpy()
```

```python
data_map
```

### Mapper


```python
# Importar: liberías
import kmapper as km
import sklearn
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Inicializar: mapper
mapper = km.KeplerMapper(verbose=1)
```

```python
# Generar: proyección sobre coordenadas
projected_data = mapper.fit_transform(data_map, projection=[2,3]) # X-Y axis
```

```python
# Generar: cubierta | segmentar méxico en 5
covering = km.Cover(n_cubes=5, perc_overlap=0.1)
```

```python
# Generar: grafo
graph = mapper.map(projected_data, data_map, clusterer=sklearn.cluster.KMeans(n_clusters=4), cover=covering)
```

```python
# Mapeo: color a base de casos confirmados
mapper.visualize(graph, path_html="Mapeo CCC.html",
                 title="Mapeo color casos confirmados",
                 color_values = data_map[:,4],
                 color_function_name = "CASOS_CONFIRMADOS",
                 node_color_function=np.array(['average','std','sum','max','min']))
```

```python
# Mapeo: color a base de población indígena
mapper.visualize(graph, path_html="Mapeo CI.html",
                 title="Mapeo color indigena",
                 color_values = data_map[:,5],
                 color_function_name = "INDIGENA",
                 node_color_function=np.array(['average','std','sum','max','min']))
```

```python
# Mapeo: color a base de población migrante
mapper.visualize(graph, path_html="Mapeo CM.html",
                 title="Mapeo color migrante",
                 color_values = data_map[:,6],
                 color_function_name = "MIGRANTE",
                 node_color_function=np.array(['average','std','sum','max','min']))
```

```python
# Estadística descriptiva: clusters
cubecluster=data.iloc[graph['nodes']['cube5_cluster1']]
cubecluster.describe()
cubecluster["ESTADO_RESIDENCIA"].value_counts()
```
