### Preparación de Datos

```python
sheet_names = [
    'SURESTE', 'NORESTE', 'CENTRO', 'NORESTE', 'SUROESTE',
    'NOROESTE2', 'NORTE', 'SUROESTE2', 'SURESTE2', 'SURESTE3',
    'SUR', 'NORTE2', 'NORESTE2', 'NORESTE3', 'NOROESTE3'
]

data_frames = {}

# Se genera un diccionario compuesto de multiples dataFrames. Cada dataFrame corresponde a los datos de una ubicación especifica.
for sheet_name in sheet_names:
    data = pd.read_excel("DATOS HISTÓRICOS 2022_2023_TODAS ESTACIONES.xlsx", sheet_name=sheet_name)
    data["ubi"] = sheet_name # Se agrega la ubicación como variable.
    data_frames[sheet_name] = data
```

```python
data_frames["SURESTE"].corr()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


for sheet in sheet_names:
    # Matriz de Correlación
    corr_matrix = data_frames[sheet].corr()

    # Dimensiones del plot
    plt.figure(figsize=(10, 8))

    # Heatmap seaborn
    sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)

    plt.title('Correlation Matrix' + sheet + '.png')

    # Save the heatmap as a PNG file
    plt.savefig('Correlation Matrix' + sheet + '.png')

    # Display the heatmap (optional)
    plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Matriz de Correlación
corr_matrix = data_frames["SURESTE2"].corr()

# Dimensiones del plot
plt.figure(figsize=(10, 8))

# Heatmap seaborn
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Se concatenan los dataFrames.
data = pd.concat(data_frames.values())
```

```python
data.to_csv("DATOS HISTÓRICOS 2022_2023_TODAS ESTACIONES CONCATENADOS.csv")
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Lista de Boxplots a generar
columns = ['CO', 'NO', 'NO2', 'O3', 'PM10', 'PM2.5', 'PRS', 'RAINF', 'RH', 'SO2', 'SR', 'TOUT', 'WSR', 'WDR']

# Diseño de la cuadricula
num_cols = 3
num_rows = (len(columns) + num_cols - 1) // num_cols

# Diseño de la Figura
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 20))
fig.tight_layout(pad=3.0) 
axes = axes.flatten()

# Generar boxplots
for i, column in enumerate(columns):
    sns.boxplot(x=data[column], ax=axes[i])
    axes[i].set_title(column)

for j in range(len(columns), num_rows * num_cols):
    axes[j].axis('off')

# Desplegar boxplots
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Lista de Columnas
columns = ['CO', 'NO', 'NO2', 'O3', 'PM10', 'PM2.5', 'PRS', 'RAINF', 'RH', 'SO2', 'SR', 'TOUT', 'WSR', 'WDR']

# Diseño de la cuadricula
num_cols = 3
num_rows = (len(columns) + num_cols - 1) // num_cols

# Ajustar el tamaño de la Figura
fig_width = 15
fig_height = num_rows * 5
fig, axes = plt.subplots(num_rows, num_cols, figsize=(fig_width, fig_height))
fig.tight_layout(pad=3.0) 
axes = axes.flatten()

# Generar histogramas
for i, column in enumerate(columns):
    sns.histplot(data[column], ax=axes[i], kde=True)  # Using histplot with KDE
    axes[i].set_title(column)

for j in range(len(columns), num_rows * num_cols):
    axes[j].axis('off')

# Desplegar histogramas
plt.show()
```
